```python
import json
from collections import defaultdict


file_path = "/content/annotations/annotations/instances_train2017.json"

with open(file_path, 'r') as f:
    data = json.load(f)
```

```python
#Constants
Info = data.get('info',{})
Liscense = data.get('licenses',{})
```

```python
#Vars
Imgs = data.get('images',[])
Anns = data.get('annotations',[])
categories = data.get('categories', [])
```

```python
#Before changing
print(f"Number of images : {len(Imgs)}")
print(f"Number of Anns : {len(Anns)}")
print(f"Number of cats : {len(categories)}")
```

```python
# 1. Number of classes

num_classes = len(categories)
print("Number of classes:", num_classes)

# Print class IDs and names
for c in categories:
    print(f"  id={c['id']}, name={c['name']:15s} , supercategory={c['supercategory']}")
```

```python
supercategory = set()
for c in categories :
  supercategory.add(c['supercategory'])

# all unique supercategory
for super in supercategory:
  print(super)
```

```python
# Number of images for each category
cat2imgs = defaultdict(set)
for ann in data.get('annotations', []):
    cat2imgs[ann['category_id']].add(ann['image_id'])
Sorted = list()


for c in categories:
    cid = c['id']
    cname = c['name']
    img_count = len(cat2imgs[cid])

    Sorted.append((cname,cid,img_count))
```

```python
Sorted = sorted(Sorted,key= lambda x : x[2],reverse=True)
print("\nImages per class:")
for i in Sorted:
  print(f"  {i[0]:15s} ({i[1]}): {i[2]} images")

```

<h2> Filteration of classes</h2>

```python
NewJson = dict()
NewJson['info'] = Info
NewJson['licenses'] = Liscense
NewJson
```

```python
Newcat= list()
removedcats = [19,11,70,5,9,14,16,20,21,22,23,24,25,32,33,34,35,36,38,39,40,41,42,43,46,56,58,59,60,61,64,74,75,76,78,79,80,81,82,85,86,87,88,89,90]
for c in categories:
  if c['id'] not in removedcats :
    Newcat.append(c)
print(f"Number of classes : {len(Newcat)}")
NewJson['categories'] = Newcat
```

```python
NewAnns = list()
UsedImgs = set()
for ann in Anns:
  if ann['category_id'] not in removedcats:
    NewAnns.append(ann)
    UsedImgs.add(ann['image_id'])
print(f"Number of anns before : {len(Anns)}")
print(f"Number of anns after : {len(NewAnns)}")
NewJson['annotations']=NewAnns
```

```python
NewImgs = list()
for img in Imgs:
  if img['id'] in UsedImgs:
    NewImgs.append(img)
print(f"Number of images before : {len(Imgs)}")
print(f"Number of images after : {len(NewImgs)}")
NewJson['images']= NewImgs
```

```python
# Number of images for each category
cat2imgs = defaultdict(set)
for ann in NewJson.get('annotations', []):
    cat2imgs[ann['category_id']].add(ann['image_id'])
Sorted = list()


for c in Newcat:
    cid = c['id']
    cname = c['name']
    img_count = len(cat2imgs[cid])

    Sorted.append((cname,cid,img_count))
    Sorted = sorted(Sorted,key= lambda x : x[2],reverse=True)
print("\nImages per class:")
for i in Sorted:
  print(f"  {i[0]:15s} ({i[1]}): {i[2]} images")

```

<h2>Choosing a randomized number of images for each class</h2>

```python
# Filteration starts
Number_Images = 3252
Total_imgs_IDs = set()
NewImgs = list()
NewAnns = list()
```

```python
"""for img in NewJson.get('images',[]):
      if img['id'] in img_ids:
        Total_imgs_IDs.add(img['id'])"""
```

```python
for c in Newcat:
  img_ids = list()#images for class
  if len(cat2imgs[c['id']]) > Number_Images :
    img_ids=list(cat2imgs[c['id']])[:Number_Images]
  else :
    img_ids=list(cat2imgs[c['id']])

  for ann in NewJson.get('annotations',[]):
      if ann['category_id'] == c['id'] and ann['image_id'] in img_ids :
        NewAnns.append(ann)
  for img in img_ids :
    Total_imgs_IDs.add(img)
```

```python
for img in NewJson.get('images',[]):
  if img['id'] in Total_imgs_IDs:
    NewImgs.append(img)

NewImgs = list (NewImgs)
NewAnns = list (NewAnns)

print(f"Number of images before : {len(NewJson.get('images',[]))}")
print(f"Number of anns before : {len(NewJson.get('annotations',[]))}")
print(f"Number of images after : {len(NewImgs)}")
print(f"Number of anns after : {len(NewAnns)}")
NewJson['images'] = NewImgs
NewJson['annotations'] = NewAnns
```

```python
# Number of images for each category
cat2imgs = defaultdict(set)
for ann in NewJson.get('annotations', []):
    cat2imgs[ann['category_id']].add(ann['image_id'])
Sorted = list()


for c in Newcat:
    cid = c['id']
    cname = c['name']
    img_count = len(cat2imgs[cid])

    Sorted.append((cname,cid,img_count))
    Sorted = sorted(Sorted,key= lambda x : x[2],reverse=True)
print("\nImages per class:")
for i in Sorted:
  print(f"  {i[0]:15s} ({i[1]}): {i[2]} images")
```

```python
with open ('filtered_train2017.json','w') as file :
  json.dump(NewJson,file)
```
